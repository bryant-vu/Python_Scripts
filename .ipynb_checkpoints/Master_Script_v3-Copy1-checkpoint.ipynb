```python
if __name__ == '__main__':
    import os, os.path
    import win32com.client
    import time
    import sys
    import pandas as pd
    import numpy
    from pathlib import Path

    #inputs from GUI
    try:
        folder_path = Path(sys.argv[1])
        report_date = sys.argv[3]
        refresh_cube = sys.argv[4]
        monthend_or_MTD = sys.argv[5]
        share_date = sys.argv[6]
        spend_date = sys.argv[7]
        forecast = sys.argv[8]
    except:
        None
    
    #file locations
    fileStringStart = folder_path / "Tools"
    fileStringEnd = str(".xlsm")
    controlWorkbook = str(folder_path / "Control Sheet - Industry Review.xlsm")
    nameplateRankWorkbook = str("Nameplate Rank Charts (M-M & YTD-YTD).xlsm")
    scorecardWorkbook = str("Health Metrics Scorecard Tool.xlsm")
    modelPerformanceWorkbook = str("Nameplate+Model Performance Slide Tool 16x9 - ")
    nameplateShareWalkWorkbook = str("Nameplate Share Walk Chart Tool - Region + Segment - ")
    regionalOverviewWorkbook = str("Nameplate+Model Regional Overview Slide Tool - ")
    segmentRankWorkbook = str("Segment Rank Charts (M-M & YTD-YTD).xlsm")
    
    #function to open workbooks
    def openWorkbook(fileLocation):
        if os.path.exists(fileLocation):
            global xl, wb
            xl=win32com.client.Dispatch("Excel.Application")
            wb = xl.Workbooks.Open(os.path.abspath(fileLocation))
            xl.Application.Visible = True   

    #function to close workbook and delete excel.exe in task manager
    def closeAndSaveWorkbook():
        wb.Close(SaveChanges=1)
        del  globals()['xl']
        time.sleep(1)

    # Common Slides
    def common_slides(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        openWorkbook(controlWorkbook)
        xl.Run("OpenPPT", str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate rank slides
        openWorkbook(str(fileStringStart / "_CommonSlides" / nameplateRankWorkbook))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("PPTAll")
        closeAndSaveWorkbook()
        #Scorecard Slides
        openWorkbook(str(fileStringStart / "_CommonSlides" / scorecardWorkbook))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPT")
        closeAndSaveWorkbook()
        #Segment rank slides
        openWorkbook(str(fileStringStart / "_CommonSlides" / segmentRankWorkbook))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("PPTAll")
        closeAndSaveWorkbook()        
        #Save CommonSlides
        openWorkbook(controlWorkbook)
        xl.Run("SaveCommonSlides", str(folder_path))
        closeAndSaveWorkbook()

    # Acura Industry Review
    def Acura_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        nameplate = "Acura"
        #Open template
        openWorkbook(controlWorkbook)
        xl.Run("OpenTemplate" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate+Model Peformance Slides
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)   
        xl.Run("NationalSlides")
        xl.Run("RegionalSlides")
        closeAndSaveWorkbook()
        #Nameplate Walk Charts Slides
        openWorkbook(str(fileStringStart / nameplate / (nameplateShareWalkWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("natCore")
        closeAndSaveWorkbook()
        #NSX Slide
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + " NSX" + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)   
        xl.Run("AutoPPTModel")
        closeAndSaveWorkbook()
        #Add Common Slides, Section Titles & Sort Slides
        openWorkbook(controlWorkbook) 
        xl.Run("CopyFromCommonSlides" + nameplate, str(folder_path))
        xl.Run("AddPPTSections" + nameplate)
        xl.Run("Sort" + nameplate)
        xl.Run("UpdateDeckCover" + nameplate)
        xl.Run("SavePPT" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
                     
    # Kia Industry Review
    def Kia_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        nameplate = "Kia"
        #Open template
        openWorkbook(controlWorkbook)
        xl.Run("OpenTemplate" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate+Model Peformance Slides
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("NationalSlides")
        xl.Run("RegionalSlides")
        closeAndSaveWorkbook()
        #Nameplate Walk Charts Slides
        openWorkbook(str(fileStringStart / nameplate / (nameplateShareWalkWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPT")
        closeAndSaveWorkbook()
        #Nameplate+Model Regional Overview Slides
        openWorkbook(str(fileStringStart / nameplate / (regionalOverviewWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPTNameplate")
        xl.Run("AutoPPTModels")
        closeAndSaveWorkbook()
        #Add Common Slides, Section Titles & Sort Slides
        openWorkbook(controlWorkbook) 
        xl.Run("CopyFromCommonSlides" + nameplate, str(folder_path))
        xl.Run("AddPPTSections" + nameplate)
        xl.Run("Sort" + nameplate)
        xl.Run("UpdateDeckCover" + nameplate)
        xl.Run("SavePPT" + nameplate, str(folder_path))
        closeAndSaveWorkbook()

    # Hyundai Industry Review
    def Hyundai_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        nameplate = "Hyundai"
        #Open template
        openWorkbook(controlWorkbook)
        xl.Run("OpenTemplate" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate+Model Peformance Slides
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("NationalSlides")
        xl.Run("RegionalSlides")
        closeAndSaveWorkbook()
        #Nameplate Walk Charts Slides
        openWorkbook(str(fileStringStart / nameplate / (nameplateShareWalkWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPT")
        closeAndSaveWorkbook()
        #Nameplate+Model Regional Overview Slides
        openWorkbook(str(fileStringStart / nameplate / (regionalOverviewWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPTNameplate")
        xl.Run("AutoPPTModels")
        closeAndSaveWorkbook()
        #Add Genesis Slides
        openWorkbook(str(fileStringStart / "Genesis" / (modelPerformanceWorkbook + "Genesis" + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("NationalSlides")
        closeAndSaveWorkbook()
        #Add Common Slides, Section Titles & Sort Slides
        openWorkbook(controlWorkbook) 
        xl.Run("CopyFromCommonSlides" + nameplate, str(folder_path))
        xl.Run("AddPPTSections" + nameplate)
        xl.Run("Sort" + nameplate)
        xl.Run("UpdateDeckCover" + nameplate)
        xl.Run("SavePPT" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
     
    # Honda Industry Review
    def Honda_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        nameplate = "Honda"
        #Open template
        openWorkbook(controlWorkbook)
        xl.Run("OpenTemplate" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate+Model Peformance Slides
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)  
        xl.Run("NationalSlides")
        xl.Run("RegionalSlides")
        closeAndSaveWorkbook()
        #Nameplate Walk Charts Slides
        openWorkbook(str(fileStringStart / nameplate / (nameplateShareWalkWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("natCore")
        closeAndSaveWorkbook()
        #Add Common Slides, Section Titles & Sort Slides
        openWorkbook(controlWorkbook) 
        xl.Run("CopyFromCommonSlides" + nameplate, str(folder_path))
        xl.Run("AddPPTSections" + nameplate)
        xl.Run("Sort" + nameplate)
        xl.Run("UpdateDeckCover" + nameplate)
        xl.Run("SavePPT" + nameplate, str(folder_path))
        closeAndSaveWorkbook()

    # Mitsubishi Industry Review
    def Mitsu_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        nameplate = "Mitsubishi"
        #Open template
        openWorkbook(controlWorkbook)
        xl.Run("OpenTemplate" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate+Model Peformance Slides
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)   
        xl.Run("NationalSlides")
        xl.Run("RegionalSlides")
        closeAndSaveWorkbook()
        #Nameplate Walk Charts Slides
        openWorkbook(str(fileStringStart / nameplate / (nameplateShareWalkWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPT")
        closeAndSaveWorkbook()
        #Nameplate+Model Regional Overview Slides
        openWorkbook(str(fileStringStart / nameplate / (regionalOverviewWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPTNameplate")
        xl.Run("AutoPPTModels")
        closeAndSaveWorkbook()
        #Add Common Slides, Section Titles & Sort Slides
        openWorkbook(controlWorkbook) 
        xl.Run("CopyFromCommonSlides" + nameplate, str(folder_path))
        xl.Run("AddPPTSections" + nameplate)
        xl.Run("Sort" + nameplate)
        xl.Run("UpdateDeckCover" + nameplate)
        xl.Run("SavePPT" + nameplate, str(folder_path))
        closeAndSaveWorkbook()

    # Mazda Industry Review
    def Mazda_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        nameplate = "Mazda"
        #Open template
        openWorkbook(controlWorkbook)
        xl.Run("OpenTemplate" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate+Model Peformance Slides
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)   
        xl.Run("NationalSlides")
        xl.Run("RegionalSlides")
        closeAndSaveWorkbook()
        #Nameplate Walk Charts Slides
        openWorkbook(str(fileStringStart / nameplate / (nameplateShareWalkWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPT")
        closeAndSaveWorkbook()
        #Nameplate+Model Regional Overview Slides
        openWorkbook(str(fileStringStart / nameplate / (regionalOverviewWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPTNameplate")
        xl.Run("AutoPPTModels")
        closeAndSaveWorkbook()
        #Add Common Slides, Section Titles & Sort Slides
        openWorkbook(controlWorkbook) 
        xl.Run("CopyFromCommonSlides" + nameplate, str(folder_path))
        xl.Run("AddPPTSections" + nameplate)
        xl.Run("Sort" + nameplate)
        xl.Run("UpdateDeckCover" + nameplate)
        xl.Run("SavePPT" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
    
    # Nissan Industry Review
    def Nissan_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        nameplate = "Nissan"
        #Open template
        openWorkbook(controlWorkbook)
        xl.Run("OpenTemplate" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate+Model Peformance Slides
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)  
        xl.Run("PowerpointDeckAll")
        closeAndSaveWorkbook()
        #Nameplate Walk Charts Slides
        openWorkbook(str(fileStringStart / nameplate / (nameplateShareWalkWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("regional")
        xl.Run("segment")
        closeAndSaveWorkbook()
        #Nameplate+Model Regional Overview Slides
        openWorkbook(str(fileStringStart / nameplate / (regionalOverviewWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPTNameplate")
        xl.Run("AutoPPTModels")
        closeAndSaveWorkbook()
        #Nameplate+Model Regional Overview Slides (No Fullsize Pickups)
        openWorkbook(str(fileStringStart / nameplate / (regionalOverviewWorkbook + nameplate + "_no_FSPU" + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPT")
        closeAndSaveWorkbook()
        #Add Common Slides, Section Titles & Sort Slides
        openWorkbook(controlWorkbook) 
        xl.Run("CopyFromCommonSlides" + nameplate, str(folder_path))
        xl.Run("AddPPTSections" + nameplate)
        xl.Run("Sort" + nameplate)
        xl.Run("UpdateDeckCover" + nameplate)
        xl.Run("SavePPT" + nameplate, str(folder_path))
        closeAndSaveWorkbook()

    # Infiniti Industry Review
    def Infiniti_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        nameplate = "Infiniti"
        #Open template
        openWorkbook(controlWorkbook)
        xl.Run("OpenTemplate" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate+Model Peformance Slides
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)  
        xl.Run("PPTAllModels")
        closeAndSaveWorkbook()
        #Add Common Slides, Section Titles & Sort Slides
        openWorkbook(controlWorkbook) 
        xl.Run("CopyFromCommonSlides" + nameplate, str(folder_path))
        xl.Run("AddPPTSections" + nameplate)
        xl.Run("Sort" + nameplate)
        xl.Run("UpdateDeckCover" + nameplate)
        xl.Run("SavePPT" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        
    #Hyundai IP
    def Hyundai_IP(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        nameplate = "Hyundai"
        controlWorkbook = str(folder_path / "Control Sheet - Incentive Planning.xlsm")
        forecastWorkbook = str(folder_path / "Forecast" / (nameplate + " Forecast.xlsm"))
        fileStringDSS = folder_path / "DSS" / nameplate
        #Open PPT
        openWorkbook(controlWorkbook)
        xl.Run("OpenPPT", str(folder_path))
        closeAndSaveWorkbook()
        #M/M & YTD/YTD Nameplate Rank Slide
        openWorkbook(str(fileStringStart / "_CommonSlides" / nameplateRankWorkbook))    
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("PPTAll")
        closeAndSaveWorkbook()
        #M/M & YTD/YTD Segment Rank Slide
        openWorkbook(str(fileStringStart / "_CommonSlides" / segmentRankWorkbook))    
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("PPTAll")
        closeAndSaveWorkbook()
        #Save CommonSlides
        openWorkbook(controlWorkbook)    
        xl.Run("SaveCommonSlides", str(folder_path))
        closeAndSaveWorkbook()
        #Open Template
        openWorkbook(controlWorkbook)    
        xl.Run("OpenTemplate" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate+Model Peformance Slides
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + " IP" + fileStringEnd)))  
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)    
        xl.Run("NationalSlides")
        xl.Run("RegionalSlides")
        closeAndSaveWorkbook()
        #Nameplate Walk Charts Slides
        openWorkbook(str(fileStringStart / nameplate / (nameplateShareWalkWorkbook + nameplate + " IP" + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)    
        xl.Run("AutoPPT")
        closeAndSaveWorkbook()
        #Nameplate+Model Regional Overview Slides
        openWorkbook(str(fileStringStart / nameplate / (regionalOverviewWorkbook + nameplate + " IP" + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)    
        xl.Run("AutoPPTNameplate")
        xl.Run("AutoPPTModels")
        closeAndSaveWorkbook()
        #Add Common Slides, Section Titles & Sort Slides
        openWorkbook(controlWorkbook)    
        xl.Run("CopyFromCommonSlides" + nameplate, str(folder_path))
        xl.Run("AddPPTSections" + nameplate)
        xl.Run("SortStandardSlides" + nameplate)
        xl.Run("UpdateDeckCover" + nameplate)
        xl.Run("SavePPT" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        
        ### Add Forecast Slides
        #Open Output PPT
        openWorkbook(controlWorkbook)    
        xl.Run("OpenOutput" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Add Forecast Slides
        openWorkbook(forecastWorkbook)    
        xl.Run("AutoPPT")
        closeAndSaveWorkbook()
        #Sort Forecast Slides
        openWorkbook(controlWorkbook)    
        xl.Run("SortForecastSlides" + nameplate)
        xl.Run("SavePPTWithForecast" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        
        ### Add DSS Slides
        #Open Output with Forecast Slides PPT
        openWorkbook(controlWorkbook)    
        xl.Run("OpenOutputWithForecast" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Grab dss file names in Control Workbook
        df = pd.read_excel(controlWorkbook, sheet_name=nameplate)
        df_slice = df.iloc[25:46,2:5]
        total_model_count = df_slice.iloc[0,0]
        fileListDSS = []
        for x in df_slice.iloc[1:,1]:
            if isinstance(x, str):
                fileListDSS.append(x)
        #Add DSS Slides
        for i in fileListDSS:
            filePath = fileStringDSS / (i + '.xlsm')
            openWorkbook(str(filePath))
            xl.Run("AutoPPTModels")
            xl.Run("AutoPPTSummary")
            closeAndSaveWorkbook()
        #Sort DSS Slides
        openWorkbook(controlWorkbook)    
        xl.Run("SortDSSSlides" + nameplate)
        xl.Run("ReOrderDSSSlides" + nameplate)
        xl.Run("SavePPTWithForecastAndDSS" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        
    #Kia IP
    def Kia_IP(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        nameplate = "Kia"
        controlWorkbook = str(folder_path / "Control Sheet - Incentive Planning.xlsm")
        forecastWorkbook = str(folder_path / "Forecast" / (nameplate + " Forecast.xlsm"))
        fileStringDSS = folder_path / "DSS" / nameplate
        #Open PPT
        openWorkbook(controlWorkbook)
        xl.Run("OpenPPT", str(folder_path))
        closeAndSaveWorkbook()
        #M/M & YTD/YTD Nameplate Rank Slide
        openWorkbook(str(fileStringStart / "_CommonSlides" / nameplateRankWorkbook))    
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("PPTAll")
        closeAndSaveWorkbook()
        #M/M & YTD/YTD Segment Rank Slide
        openWorkbook(str(fileStringStart / "_CommonSlides" / segmentRankWorkbook))    
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("PPTAll")
        closeAndSaveWorkbook() 
        #Save CommonSlides
        openWorkbook(controlWorkbook)    
        xl.Run("SaveCommonSlides", str(folder_path))
        closeAndSaveWorkbook()
        #Open Template
        openWorkbook(controlWorkbook)    
        xl.Run("OpenTemplate" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate+Model Peformance Slides
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + " IP" + fileStringEnd)))    
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)    
        xl.Run("NationalSlides")
        xl.Run("RegionalSlides")
        closeAndSaveWorkbook()
        #Nameplate Walk Charts Slides
        openWorkbook(str(fileStringStart / nameplate / (nameplateShareWalkWorkbook + nameplate + " IP" + fileStringEnd)))  
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)    
        xl.Run("AutoPPT")
        closeAndSaveWorkbook()
        #Nameplate+Model Regional Overview Slides
        openWorkbook(str(fileStringStart / nameplate / (regionalOverviewWorkbook + nameplate + " IP" + fileStringEnd)))   
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)    
        xl.Run("AutoPPTNameplate")
        xl.Run("AutoPPTModels")
        closeAndSaveWorkbook()
        #Add Common Slides, Section Titles & Sort Slides
        openWorkbook(controlWorkbook)    
        xl.Run("CopyFromCommonSlides" + nameplate, str(folder_path))
        xl.Run("AddPPTSections" + nameplate)
        xl.Run("SortStandardSlides" + nameplate)
        xl.Run("UpdateDeckCover" + nameplate)
        xl.Run("SavePPT" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        
        ### Add Forecast Slides
        #Open Output PPT
        openWorkbook(controlWorkbook)    
        xl.Run("OpenOutput" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Add Forecast Slides
        openWorkbook(forecastWorkbook)    
        xl.Run("AutoPPT")
        closeAndSaveWorkbook()
        #Sort Forecast Slides
        openWorkbook(controlWorkbook)    
        xl.Run("SortForecastSlides" + nameplate)
        xl.Run("SavePPTWithForecast" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        
        ### Add DSS Slides
        #Open Output with Forecast Slides PPT
        openWorkbook(controlWorkbook)    
        xl.Run("OpenOutputWithForecast" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Grab dss file names in Control Workbook
        df = pd.read_excel(controlWorkbook, sheet_name=nameplate)
        df_slice = df.iloc[25:46,2:5]
        total_model_count = df_slice.iloc[0,0]
        fileListDSS = []
        for x in df_slice.iloc[1:,1]:
            if isinstance(x, str):
                fileListDSS.append(x)
        #Add DSS Slides
        for i in fileListDSS:
            filePath = fileStringDSS / (i + '.xlsm')
            openWorkbook(str(filePath))
            xl.Run("AutoPPTModels")
            xl.Run("AutoPPTSummary")
            closeAndSaveWorkbook()
        #Sort DSS Slides
        openWorkbook(controlWorkbook)    
        xl.Run("SortDSSSlides" + nameplate)
        xl.Run("ReOrderDSSSlides" + nameplate)
        xl.Run("SavePPTWithForecastAndDSS" + nameplate, str(folder_path))
        closeAndSaveWorkbook()

    def Mitsu_IP(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        nameplate = "Mitsubishi"
        controlWorkbook = str(folder_path / "Control Sheet - Incentive Planning.xlsm")
        forecastWorkbook = str(folder_path / "Forecast" / (nameplate + " Forecast.xlsm"))
        fileStringDSS = folder_path / "DSS" / nameplate
        #Open PPT
        openWorkbook(controlWorkbook)    
        xl.Run("OpenPPT", str(folder_path))
        closeAndSaveWorkbook()
        #M/M & YTD/YTD Nameplate Rank Slide
        openWorkbook(str(fileStringStart / "_CommonSlides" / nameplateRankWorkbook))    
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("PPTAll")
        closeAndSaveWorkbook()
        #M/M & YTD/YTD Segment Rank Slide
        openWorkbook(str(fileStringStart / "_CommonSlides" / segmentRankWorkbook))    
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("PPTAll")
        closeAndSaveWorkbook() 
        #Save CommonSlides
        openWorkbook(controlWorkbook)    
        xl.Run("SaveCommonSlides", str(folder_path))
        closeAndSaveWorkbook()
        #Open Template
        openWorkbook(controlWorkbook)    
        xl.Run("OpenTemplate" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate+Model Peformance Slides
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("NationalSlides")
        closeAndSaveWorkbook()
        #Nameplate Walk Charts Slides
        openWorkbook(str(fileStringStart / nameplate / (nameplateShareWalkWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("natCore")
        xl.Run("regional")
        closeAndSaveWorkbook()
        #Add Common Slides, Section Titles & Sort Slides
        openWorkbook(controlWorkbook)    
        xl.Run("CopyFromCommonSlides" + nameplate, str(folder_path))
        xl.Run("AddPPTSections" + nameplate)
        xl.Run("SortStandardSlides" + nameplate)
        xl.Run("UpdateDeckCover" + nameplate)
        xl.Run("SavePPT" + nameplate, str(folder_path))
        closeAndSaveWorkbook()

    def Nissan_IP(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast):
        nameplate = "Nissan"
        controlWorkbook = str(folder_path / "Control Sheet - Incentive Planning.xlsm")
        forecastWorkbook = str(folder_path / "Forecast" / (nameplate + " Forecast.xlsm"))
        fileStringDSS = folder_path / "DSS" / nameplate
        #Open PPT
        openWorkbook(controlWorkbook)    
        xl.Run("OpenPPT", str(folder_path))
        closeAndSaveWorkbook()
        #M/M & YTD/YTD Nameplate Rank Slide
        openWorkbook(str(fileStringStart / "_CommonSlides" / nameplateRankWorkbook))    
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("PPTAll")
        closeAndSaveWorkbook()
        #M/M & YTD/YTD Segment Rank Slide
        openWorkbook(str(fileStringStart / "_CommonSlides" / segmentRankWorkbook))    
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("PPTAll")
        closeAndSaveWorkbook() 
        #Save CommonSlides
        openWorkbook(controlWorkbook)    
        xl.Run("SaveCommonSlides", str(folder_path))
        closeAndSaveWorkbook()       
        #Open template
        openWorkbook(controlWorkbook)
        xl.Run("OpenTemplate" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        #Nameplate+Model Peformance Slides
        openWorkbook(str(fileStringStart / nameplate / (modelPerformanceWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)  
        xl.Run("PowerpointDeckAll")
        closeAndSaveWorkbook()
        #Nameplate Walk Charts Slides
        openWorkbook(str(fileStringStart / nameplate / (nameplateShareWalkWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("regional")
        xl.Run("segment")
        closeAndSaveWorkbook()
        #Nameplate+Model Regional Overview Slides
        openWorkbook(str(fileStringStart / nameplate / (regionalOverviewWorkbook + nameplate + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPTNameplate")
        xl.Run("AutoPPTModels")
        closeAndSaveWorkbook()
        #Nameplate+Model Regional Overview Slides (No Fullsize Pickups)
        openWorkbook(str(fileStringStart / nameplate / (regionalOverviewWorkbook + nameplate + "_no_FSPU" + fileStringEnd)))
        xl.Run("update_slicer", report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        xl.Run("AutoPPT")
        closeAndSaveWorkbook()
        #Add Common Slides, Section Titles & Sort Slides
        openWorkbook(controlWorkbook) 
        xl.Run("CopyFromCommonSlides" + nameplate, str(folder_path))
        xl.Run("AddPPTSections" + nameplate)
        xl.Run("SortStandardSlides" + nameplate)
        xl.Run("UpdateDeckCover" + nameplate)
        xl.Run("SavePPT" + nameplate, str(folder_path))
        closeAndSaveWorkbook()
        
    #Run functions
    if sys.argv[2] == 'Industry_Review_(all)':
        common_slides(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Acura_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Honda_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Hyundai_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Kia_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Mazda_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Mitsu_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Nissan_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Infiniti_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Incentive_Planning_(all)':
        common_slides(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Kia_IP(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Hyundai_IP(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Mitsu_IP(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Nissan_IP(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Acura_IR':
        common_slides(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Acura_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Honda_IR':
        common_slides(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Honda_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Hyundai_IR':
        common_slides(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Hyundai_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Kia_IR':
        common_slides(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Kia_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Mazda_IR':
        common_slides(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Mazda_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Mitsu_IR':
        common_slides(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Mitsu_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Nissan_IR':
        common_slides(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Nissan_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Infiniti_IR':
        common_slides(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
        Infiniti_IR(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Hyundai_IP':
        Hyundai_IP(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Kia_IP':
        Kia_IP(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Mitsu_IP':
        Mitsu_IP(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
    elif sys.argv[2] == 'Nissan_IP':
        Nissan_IP(folder_path, report_date, refresh_cube, monthend_or_MTD, share_date, spend_date, forecast)
```
