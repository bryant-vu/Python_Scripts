```python
import PySimpleGUI as sg
import os
from Scenario_Chooser_Main_v2 import *
from pathlib import Path

exit = False
# Window #1 (Scenario_Chooser folder path)
layout = [[sg.Text('Path to Scenario_Chooser folder:')],
          [sg.Listbox(values=('C:/Users/jimmy.gov/Desktop/Scenario_Chooser', 'C:/Users/bryant.vu/Documents/Scenario_Chooser'), size=(80, 5))],
          [sg.OK(), sg.Cancel()]]
window = sg.Window('Scenario Chooser', layout)

while True:
    event, values = window.Read()
    if event is None or event == 'Exit' or event == 'Cancel':
        exit = True
        break
    else:
        DSS_folder_path = values[0][0] + "/DSS_files"
        gif_file_path = values[0][0] + "/.gif Files/monkey.gif"
        DSS_files_list = os.listdir(DSS_folder_path)
        DSS_files_list = [x for x in DSS_files_list if ".xlsm" in x]
        break
window.Close()

while(exit==False):
    # Window #2 (DSS file path)
    layout = [[sg.Text('DSS file:')],
              [sg.Listbox(values=DSS_files_list, size=(80, 10))],
              [sg.Text('New or old version?'), sg.Combo(['New', 'Old'], default_value = 'New', size=(5,1))],
              [sg.OK(), sg.Cancel()]]
    window = sg.Window('Scenario Chooser', layout)

    while True:
        event, values = window.Read()
        if event is None or event == 'Exit' or event == 'Cancel':
            exit = True
            break
        else:
            DSS_file_path = DSS_folder_path + "/" + values[0][0]
        new_or_old_version = values[1]
        if new_or_old_version == 'New':
            model_list, output_tabs  = DSS_extract_models_and_output_tabs_new_DSS(DSS_file_path)
            break
        else:
            model_list, output_tabs  = DSS_extract_models_and_output_tabs_old_DSS(DSS_file_path)
            break
    window.Close()
    if exit==True:
        break

    # Window 3 (user options)
    layout = [[sg.Text('Choose model:')],
              [sg.Listbox(values=model_list, size=(80, 10))],
              [sg.Text('Choose output tab:')],
              [sg.Listbox(values=output_tabs, size=(80, 10))],
              [sg.Text('Minimum spend ($):'), sg.InputText()],
              [sg.Text('Maximum spend ($):'), sg.InputText()],
              [sg.Text('Max # of scenarios:'), sg.InputText()],
              [sg.Text('Optional: Include the lowest incremental single lever move (or 2x)')],
              [sg.Checkbox('CC'), sg.Checkbox('2x CC'), sg.Checkbox('-1ppt (all terms)'), sg.Checkbox('Combo'),sg.Checkbox('2x Combo'), sg.Checkbox('Lease'), sg.Checkbox('2x Lease')],
              [sg.Text('Additional filters')],
              [sg.Checkbox('Remove APR scenarios with \'std\'')],
              [sg.Checkbox('[Mitsu] Show same Combo Cash move with CC move')],
              [sg.OK(), sg.Cancel(), sg.Button('Back')]]
    window = sg.Window('Scenario Chooser', layout)

    while True:
        event, values = window.Read()
        if event is None or event == 'Exit' or event == 'Cancel':
            exit = True
            break
        elif event == 'Back':
            break
        elif (not bool(values[0])) or (not bool(values[1])) or (not bool(values[2])) or (not bool(values[3])) or (not bool(values[4])):
            sg.Popup('Required inputs: model, output tab, min_spend, max_spend, max # of scenarios')
            continue
        else:
            inputs_list = ['model', 'output_tab', 'min_spend', 'max_spend','max_no_of_scenarios', 'min_cash_bool', 'dbl_min_cash_bool', 'APR_bool', 'min_combo_bool', 'dbl_min_combo_bool', 'min_lease_bool', 'dbl_min_lease_bool', 'std_bool', 'CC_combo_bool']
        inputs_dict = {}
        for i, input in enumerate(inputs_list):
            if i == 0 or i == 1:
                inputs_dict[input] = values[i][0]
            elif i >= 2 or i <= 4 : 
                inputs_dict[input] = int(values[i])
            else:
                inputs_dict[input] = values[i]
        inputs_dict['model_order'] = model_list.index(inputs_dict['model']) + 1 
        sg.PopupAnimated(gif_file_path, message='Just a sec...', background_color='#FFF', font=("Helvetica",18), keep_on_top=True, location=(0,0))     

        #New or old DSS
        if new_or_old_version == 'New':
            run_chooser_new_DSS(DSS_file_path, inputs_dict['output_tab'], inputs_dict['model_order'], inputs_dict['min_cash_bool'], inputs_dict['dbl_min_cash_bool'],inputs_dict['APR_bool'], inputs_dict['min_combo_bool'], inputs_dict['dbl_min_combo_bool'], inputs_dict['min_lease_bool'], inputs_dict['dbl_min_lease_bool'], inputs_dict['std_bool'], inputs_dict['CC_combo_bool'], inputs_dict['min_spend'], inputs_dict['max_spend'], inputs_dict['max_no_of_scenarios'])
        else:
             run_chooser_old_DSS(DSS_file_path, inputs_dict['output_tab'], inputs_dict['model_order'], inputs_dict['min_cash_bool'], inputs_dict['dbl_min_cash_bool'],inputs_dict['APR_bool'], inputs_dict['min_combo_bool'], inputs_dict['dbl_min_combo_bool'], inputs_dict['min_lease_bool'], inputs_dict['dbl_min_lease_bool'], inputs_dict['std_bool'], inputs_dict['CC_combo_bool'], inputs_dict['min_spend'], inputs_dict['max_spend'], inputs_dict['max_no_of_scenarios'])
        sg.PopupAnimated(image_source=None)
        sg.Popup("Done.", keep_on_top=True)       
    window.Close()
    if exit==True:
        break
```
